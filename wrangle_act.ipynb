#### Tidiness 

 ####  Define
 - using replace  in different  dog stages to remove 'None',merging the dog stages columns,managing multiple stages and missing values.

#### Code

```python

# handle none
df_enhanced.doggo.replace('None', '', inplace=True)
df_enhanced.doggo.replace('np.NaN', '', inplace=True)
df_enhanced.floofer.replace('None', '', inplace=True)
df_enhanced.floofer.replace('np.NaN', '', inplace=True)
df_enhanced.pupper.replace('None', '', inplace=True)
df_enhanced.pupper.replace('np.NaN', '', inplace=True)
df_enhanced.puppo.replace('None', '', inplace=True)
df_enhanced.puppo.replace('np.NaN', '', inplace=True)
# merge into column
df_enhanced['dog_stage'] = df_enhanced.doggo + df_enhanced.floofer + df_enhanced.pupper + df_enhanced.puppo
# handle multiple stages
df_enhanced.loc[df_enhanced.dog_stage == 'doggopupper', 'dog_stage'] = 'doggo, pupper'
df_enhanced.loc[df_enhanced.dog_stage == 'doggopuppo', 'dog_stage'] = 'doggo, puppo'
df_enhanced.loc[df_enhanced.dog_stage == 'doggofloofer', 'dog_stage'] = 'doggo, floofer'
```

#### Define
- creating  a p_dogs  columns  from  the prediction of breeds  

#### Code

```python
dog_prediction = []
for i in range(2075):
    if df_image['p1_dog'][i]:
        dog_prediction.append("p1_dog")
    elif df_image['p2_dog'][i]:
         dog_prediction.append("p2_dog")
    elif df_image['p3_dog'][i]:
         dog_prediction.append("p3_dog")
    else:
        dog_prediction.append("no prediction")
# creating new column from above array
df_image['p_dogs'] = dog_prediction
```
